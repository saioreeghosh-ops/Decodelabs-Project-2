```python
import pandas as pd
```

```python
df = pd.read_excel("cleaned_data.xlsx")
```

```python
print(df.head())
```

```python
import matplotlib.pyplot as plt
```

```python
# Correlation between numerical columns
print(df.corr(numeric_only=True))
```

```python
df.groupby("Product")["TotalPrice"].sum().plot(kind="bar")

plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
df["TotalPrice"].plot(kind="hist", bins=10)

plt.title("Distribution of TotalPrice")
plt.xlabel("TotalPrice")
plt.show()
```

```python
# Top 5 highest sales
top_sales = df.sort_values(by="TotalPrice", ascending=False)

print(top_sales.head())
```

```python
# Average sales per product
avg_sales = df.groupby("Product")["TotalPrice"].mean()

print(avg_sales)
```

```python
# Detect outliers using IQR
Q1 = df["TotalPrice"].quantile(0.25)
Q3 = df["TotalPrice"].quantile(0.75)

IQR = Q3 - Q1

outliers = df[(df["TotalPrice"] < (Q1 - 1.5 * IQR)) | 
              (df["TotalPrice"] > (Q3 + 1.5 * IQR))]

print(outliers)
```

```python
plt.boxplot(df["TotalPrice"])

plt.title("Boxplot of TotalPrice")
plt.show()
```

```python
df["Product"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.title("Product Distribution")
plt.ylabel("")
plt.show()
```
